```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.currency-converter.org.uk/currency-rates/historical/table/EUR-USD.html"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup.select("#content > table")
```

```python
table = soup.select("#content > table")[0]
```

```python
table
```

```python
import pandas as pd
table_rows = table.find_all('tr')


l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)

df =pd.DataFrame(l, columns=["Week Day", "Date", "Euro","Dollar" , "Analysis"])
```

```python
df.head()
```

```python
df.drop([0])
```

```python
df = df.drop([0])
```

```python
df = df.drop(columns=["Week Day", "Analysis"])
```

```python
df.head()
```

```python
df['Dollar'] = df['Dollar'].str.replace('USD', '')
```

```python
df.head()
```

```python
df['Euro'] = df['Euro'].str.replace('EUR =', '')
```

```python
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"]).dt.strftime('%m/%d/%Y')
```

```python
df.head()
```

```python
df['Euro'] = df['Euro'].astype(str).astype(int)
```

```python
df['Dollar'] = df['Dollar'].astype(str).astype(float)
```

```python
df.info()
```

```python
df.to_csv("currency.csv")
```
